```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import the dataset
df = pd.read_csv('grass_growth_dataset.csv')

# Create the figure
plt.figure(figsize=(12, 6))

# Plot the linear regression
sns.regplot(x='RAIN', y='GROWTH', data=df, scatter_kws={'alpha':0.5, 'color':'blue'}, line_kws={'color':'red'}, ci=None)

# Plot the second-order (quadratic) regression
sns.regplot(x='RAIN', y='GROWTH', data=df, scatter=False, order=2, line_kws={'color':'green', 'linestyle':'--'})

plt.title('Regression Plot of Grass Growth vs. Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Grass Growth (units)')
plt.grid(True)
plt.legend(['data', 'Linear Regression', 'Polynomial Regression'])
plt.show()
```

# Coding a polynomial regression model with scikit-learn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 30)
y = [3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179]
plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.show()
```

## STEP #1: Determining the degree of the polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
```

degree sets the degree of our polynomial function. degree=2 means that we want to work with a 2nd degree polynomial:

y = 0 + 1x + 2x2

include_bias=False should be set to False, because well use PolynomialFeatures together with LinearRegression() later on.

## STEP #2: Creating the new features

```python
poly_features = poly.fit_transform(x.reshape(-1, 1))
```

```python
poly_features
```

reshape(-1,1) transforms our numpy array x from a 1D array to a 2D array  this is required, otherwise wed get the following error:

valueerror expected 2d array

## STEP #3: Creating the polynomial regression model

```python
from sklearn.linear_model import LinearRegression
poly_reg_model = LinearRegression()
```

```python
poly_reg_model.fit(poly_features, y)
```

```python
y_predicted = poly_reg_model.predict(poly_features)
```

```python
plt.figure(figsize=(10, 6))
plt.title("Your first polynomial regression  congrats! :)", size=16)
plt.scatter(x, y)
plt.plot(x, y_predicted, c="red")
plt.show()
```
